```python
#Importing the dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import seaborn as sns
```

```python
#Uploading the datasets
movies = pd.read_csv(r'G:\Projects\Movie Recomendation Analysis\Movies.csv')
more_info = pd.read_csv(r'G:\Projects\Movie Recomendation Analysis\MoreInfo.csv')
film_details = pd.read_csv(r'G:\Projects\Movie Recomendation Analysis\FilmDetails.csv')
poster_path = pd.read_csv(r'G:\Projects\Movie Recomendation Analysis\PosterPath.csv')
```

```python
#Merging the datasets together
watch = pd.merge(movies, film_details, on = 'id', how = 'inner')
watch = pd.merge(watch, more_info, on = 'id', how = 'inner')
watch = pd.merge(watch, poster_path, on = 'id', how = 'inner')

watch.to_csv('watch_movies.csv', index= False)
print("All datasets have been merged successfully into 'watch_movies.csv'")
```

```python
#Importing the new dataset
watch = pd.read_csv(r'G:\Projects\Movie Recomendation Analysis\watch_movies.csv')
```

```python
#Seeing the first five columns of the dataset
watch.head()
```

```python
#As there is revenue and budget which we don't need. So, lets drop those columns
watch = watch.drop(['revenue', 'film_id', 'budget'], axis=1)
```

```python
#Now, lets confirm if the columns are dropped
watch.head()
```

```python
#Now let's analyse the data
watch.describe()
```

```python
#Checking if the dataset have some null value
print(watch.isnull().sum())

#Deleting or filling the null values
watch['budget_usd'] = watch['budget_usd'].fillna(0)
watch['vote_count'] = watch['vote_count'].fillna(0)
watch['genres'] = watch['genres'].fillna('Unknown')
watch['top_billed'] = watch['top_billed'].fillna('Unknown')
```

```python
#Creating a distribution of movie budgets
sns.histplot(watch['budget_usd'], kde= True, bins= 30, color = 'purple')

plt.title('Distribution of Movie Budgets')
plt.show
```

```python
#Count the movies my genres
genre_counts = watch['genres'].value_counts()

# Plot top 10 genres
genre_counts.head(10).plot(kind='barh', figsize=(10, 5), color = 'purple')
plt.title("Top 10 Genres by Movie Count")
plt.ylabel("Genres")
plt.xlabel("Number of Movies")
plt.show()
```

```python
print(watch['vote_count'].describe())


sns.scatterplot(x = 'budget_usd', y = 'vote_count', data= watch, color = 'purple')
plt.title('Budget vs Vote Counts')
plt.show
```

```python
actor_counts = watch['top_billed'].value_counts()

actor_counts.head(10).plot(kind = 'barh', figsize = (10, 5), color = 'purple')
plt.title('Top 10 Billed Actors By Movie Counts')
plt.xlabel('Movies')
plt.ylabel('Actors')
```
